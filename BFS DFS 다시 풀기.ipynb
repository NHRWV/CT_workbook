```python
# 백준 2178 미로 탐색

from collections import deque

n, m = map(int, input().split())
maze = []
for i in range(n):
    maze.append(list(map(int, input())))
    
queue = deque([(0, 0)])
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

while queue:
    x, y = queue.popleft()
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
    
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
            
        
        if maze[nx][ny] == 1:
            maze[nx][ny] = maze[x][y] + 1
            queue.append((nx, ny))
            
            
print(maze[n-1][m-1])
    
```

https://www.acmicpc.net/problem/2606

```python
# 백준 2606 바이러스

from collections import deque

n = int(input())
m = int(input())

net = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    net[a].append(b)
    net[b].append(a)
    
queue = deque([1])
visited = [0] * (n+1)
visited[1] = 1
count = 0
while queue:
    x = queue.popleft()

    for i in net[x]:
        if not visited[i]:
            visited[i] = 1
            queue.append(i)
            count += 1
            
            
print(count)
```

```python
# 백준 7576 토마토

from collections import deque
import sys

m, n = map(int, input().split())

tomato = []
ripen = []
check = 0
for i in range(n):
    a = list(map(int, input().split()))
    tomato.append(a)
    for j in range(m):
        if a[j] == 1:
            ripen.append((i, j))
        
        elif a[j] == 0:
            check = 1
            
if check == 0:
    print(0)
    sys.exit()
            
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
            
queue = deque(ripen)

while queue:
    x, y = queue.popleft()
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
            
        if tomato[nx][ny] == 0:
            tomato[nx][ny] = tomato[x][y] + 1
            queue.append((nx, ny))
            
            
answer = 0
for i in range(n):
    answer = max(answer, max(tomato[i]))
    if tomato[i].count(0) > 0:
        print(-1)
        sys.exit()
        
print(answer - 1)
```

```python
# 백준 1012 유기농 배추

from collections import deque

def sol(x, y):
    queue = deque([(x, y)])
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    
    while queue:
        x, y = queue.popleft()
        visited[x][y] = 1
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
                
            if farm[nx][ny] == 1 and not visited[nx][ny]:
                visited[nx][ny] = 1
                queue.append((nx, ny))


t = int(input())
for i in range(t):
    m, n, k = map(int, input().split())
    
    farm = [[0] * m for _ in range(n)]
    for j in range(k):
        a, b = map(int, input().split())
        farm[b][a] = 1
        
    visited = [[0] * m for _ in range(n)]
    count = 0
    for p in range(n):
        for q in range(m):
            if not visited[p][q] and farm[p][q] == 1:
                sol(p, q)
                count += 1
                
                
    print(count)
```

https://www.acmicpc.net/problem/14502

```python
# 백준 14502 연구소

from collections import deque
from itertools import combinations
from copy import deepcopy


n, m = map(int, input().split())

lab = []
empty = []
virus = []
for i in range(n):
    a = list(map(int, input().split()))
    lab.append(a)
    for j in range(m):
        if a[j] == 0:
            empty.append((i, j))
        
        elif a[j] == 2:
            virus.append((i, j))
            

walls =  list(combinations(empty, 3))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

answer = 0
for wall in walls:
    copy = deepcopy(lab)
    for a, b in wall:
        copy[a][b] = 1
        
    
    queue = deque(virus)
    
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
                
            if copy[nx][ny] == 0:
                copy[nx][ny] = 2
                queue.append((nx, ny))
                
                
    count = 0
    for i in range(n):
        for j in range(m):
            if copy[i][j] == 0:
                count += 1
                
    answer = max(answer, count)
    
    
    
print(answer)
```

```python
# 백준 2644 촌수 계산

from collections import deque

n = int(input())
t1, t2 = map(int, input().split())
m = int(input())
edge = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    edge[a].append(b)
    edge[b].append(a)
    
    
visited = [-1] * (n+1)
visited[t1] = 0
queue = deque([t1])

while queue:
    x = queue.popleft()
    
    for i in edge[x]:
        if visited[i] == -1:
            visited[i] = visited[x] + 1
            queue.append(i)
            
            
if visited[t2] == -1:
    print(-1)
    
else:
    print(visited[t2])
```
