```python
# 백준 16946 벽 부수고 이동하기 4

from collections import deque
from copy import deepcopy

n, m = map(int, input().split())

space = []
wall = []
empty = []
for i in range(n):
    a = list(map(int, input().rstrip()))
    space.append(a)
    for j in range(m):
        if a[j] == 1:
            wall.append((i, j))
            
        elif a[j] == 0:
            empty.append((i, j))
    
    
    
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
copy = deepcopy(space)

def numbering():
    group = 0
    visited = [[0] * m for _ in range(n)]
    
    for a, b in empty:
        if visited[a][b]:
            continue
            
        visited[a][b] = 1
        queue = deque([(a, b)])
        lst = [(a, b)]
        
        while queue:
            x, y = queue.popleft()
            
            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]
                
                if nx < 0 or nx >= n or ny < 0 or ny >= m:
                    continue
                    
                if not visited[nx][ny] and not space[nx][ny]:
                    visited[nx][ny] = 1
                    queue.append((nx, ny))
                    lst.append((nx, ny))
        
        cnt = len(lst)
        for i in lst:
            space[i[0]][i[1]] = (cnt, group)

        group += 1
        
        
        

def counting():
    
    
    for a, b in wall:
        count = 1
        censor = set()
        
        for i in range(4):
            na = a + dx[i]
            nb = b + dy[i]
            
            if na < 0 or na >= n or nb < 0 or nb >= m:
                continue
                
            if space[na][nb] != 1 and space[na][nb] not in censor:
                count += space[na][nb][0]
                censor.add(space[na][nb])
                
        copy[a][b] = count % 10
        
    
    
    
numbering()
counting()

for i in copy:
    for j in i:
        print(j, end = '')
    print()
    
    
```

```python
from collections import deque
from copy import deepcopy
from itertools import combinations


n = int(input())


teacher = []
empty = []
area = []
for i in range(n):
    a = list(map(str, input().split()))
    area.append(a)
    for j in range(n):
        if a[j] == 'T':
            teacher.append((i, j))
        elif a[j] == 'X':
            empty.append((i, j))
            
lst = list(combinations(empty, 3))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def sol():
    def check(x, y, dx, dy):
        while True:
            nx = x + dx
            ny = y + dy
            
            if nx < 0 or nx >= n or ny < 0 or ny >= n:
                return True
                
            if copy[nx][ny] == 'S':
                return False
            
            if copy[nx][ny] == 'O':
                return True
            
            x += dx
            y += dy
        
    
    for combi in lst:
        copy = deepcopy(area)
        flag = 1
        
        for i in combi:
            copy[i[0]][i[1]] = 'O'
            
        queue = deque(teacher)
        
        while queue:
            x, y = queue.popleft()
            
            for i in range(4):
                if not check(x, y, dx[i], dy[i]):
                    flag = 0
                    break
                    
        if flag:
            return 'YES'
                          
    return "NO"


print(sol())
            
        
    
```

```python
from collections import deque

n, m, k, x = map(int, input().split())

edge = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    edge[a].append(b)
    
    
visited = [-1] * (n+1)
visited[x] = 0

queue = deque([x])
while queue:
    x = queue.popleft()
    
    for i in edge[x]:
        if visited[i] == -1:
            visited[i] = visited[x] + 1
            queue.append(i)
            

if k not in set(visited):
    print(-1)
    
else:
    for i, j in enumerate(visited):
        if j == k:
            print(i)
            
```
