```python
# 결혼식

from collections import deque

n = int(input())
m = int(input())

edge = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    edge[a].append(b)
    edge[b].append(a)
    
queue = deque([1])

visited = [0] * (n+1)

while queue:
    x = queue.popleft()
    
    for i in edge[x]:
        if not visited[i]:
            visited[i] = visited[x] + 1
            queue.append(i)
            
            
print(visited[2:].count(1) + visited[2:].count(2))
```

```python
# 양치기 꿍

from collections import deque

r, c = map(int, input().split())

area = []
for i in range(r):
    area.append(list(map(str, input())))
    
    
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
visited = [[0] * c for _ in range(r)]

wolf = 0
sheep = 0

def sol(x, y):
    global wolf, sheep
    
    visited[x][y] = 1
    w = 0
    s = 0
    
    if area[x][y] == 'v':
        w += 1
    if area[x][y] == 'k':
        s += 1
    
    queue = deque([(x, y)])
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= r or ny < 0 or ny >= c:
                continue
            
            if not visited[nx][ny] and area[nx][ny] != '#':
                visited[nx][ny] = 1
                
                if area[nx][ny] == 'v':
                    w += 1
                if area[nx][ny] == 'k':
                    s += 1
                    
                queue.append((nx, ny))
                
    if w < s:
        sheep += s
    else:
        wolf += w
        
        
        
for i in range(r):
    for j in range(c):
        if not visited[i][j] and area[i][j] != '#':
            sol(i, j)
            
            
print(sheep, wolf)
            
                    
    
```
