```python
# LCA


def meas_level(x, depth):
    visited[x] = 1
    level[x] = depth
    
    for i in edge[x]:
        if not visited[i]:
            parent[i] = x
            meas_level(i, depth+1)
    
    
def lca(a, b):
    while level[a] != level[b]:
        if level[a] < level[b]:
            b = parent[b]
            
        else:
            a = parent[a]
            
    while a != b:
        a = parent[a]
        b = parent[b]
        
        
    return a


n = int(input())

edge = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b = map(int, input().split())
    edge[a].append(b)
    edge[b].append(a)
    
    
visited = [0] * (n+1)
parent = [0] * (n+1)
level = [0] * (n+1)

meas_level(1, 0)

m = int(input())
for i in range(m):
    a, b = map(int, input().split())
    
    print(lca(a, b))
```

```python
# LCA 2


def meas_level(x, depth):
    visited[x] = 1
    level[x] = depth
    
    for i in edge[x]:
        if not visited[i]:
            parent[i][0] = x
            meas_level(i, depth+1)
    
    
def lca(a, b):
    if level[b] < level[a]:
        a, b = b, a
        
    for i in range(17, -1, -1):
        if level[b] - level[a] >= 2 ** i:
            b = parent[b][i]
            
    if a == b:
        return a
    
    for i in range(17, -1, -1):
        if parent[a][i] != parent[b][i]:
            a = parent[a][i]
            b = parent[b][i]
        
        
    return parent[a][0]


n = int(input())

edge = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b = map(int, input().split())
    edge[a].append(b)
    edge[b].append(a)
    
    
visited = [0] * (n+1)
level = [0] * (n+1)
parent = [[0] * 18 for _ in range(n+1)] # log(2,100000)


meas_level(1, 0)

for i in range(1, 18): # 노드의 부모 관계 설정
    for j in range(1, n+1):
        parent[j][i] = parent[parent[j][i-1]][i-1]
        
        
m = int(input())
for i in range(m):
    a, b = map(int, input().split())
    print(lca(a, b))
    
```

https://www.acmicpc.net/problem/1761

```python
# 정점들의 거리

def meas_level(x, senior, dist):
    visited[x] = 1
    level[x] = level[senior] + 1
    
    if x != 1:
        distance[x] += distance[senior] + dist
    
    for i, j in edge[x]:
        if not visited[i]:
            parent[i][0] = x
            meas_level(i, x, j)
    
    
def lca(a, b):
    if level[b] < level[a]:
        a, b = b, a
        
    for i in range(17, -1, -1):
        if level[b] - level[a] >= 2 ** i:
            b = parent[b][i]
            
    if a == b:
        return a
    
    for i in range(17, -1, -1):
        if parent[a][i] != parent[b][i]:
            a = parent[a][i]
            b = parent[b][i]
        
        
    return parent[a][0]


n = int(input())

edge = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b, c = map(int, input().split())
    edge[a].append((b, c))
    edge[b].append((a, c))
    
    
visited = [0] * (n+1)
parent = [[0] * 18 for _ in range(n+1)] # log(2,100000)

distance = [0] * (n+1)
level = [0] * (n+1)
level[0] = -1

meas_level(1, 0, 0)

for i in range(1, 18): # 노드의 부모 관계 설정
    for j in range(1, n+1):
        parent[j][i] = parent[parent[j][i-1]][i-1]
        
        
m = int(input())
for i in range(m):
    a, b = map(int, input().split())
    # 최단거리 = distance[a] - distance[lca] + distance[b] - distance[lca]
    print(distance[a] + distance[b] - 2 * distance[lca(a, b)])
    
```
