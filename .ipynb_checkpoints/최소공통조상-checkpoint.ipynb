```python
# LCA


def meas_level(x, depth):
    visited[x] = 1
    level[x] = depth
    
    for i in edge[x]:
        if not visited[i]:
            parent[i] = x
            meas_level(i, depth+1)
    
    
def lca(a, b):
    while level[a] != level[b]:
        if level[a] < level[b]:
            b = parent[b]
            
        else:
            a = parent[a]
            
    while a != b:
        a = parent[a]
        b = parent[b]
        
        
    return a


n = int(input())

edge = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b = map(int, input().split())
    edge[a].append(b)
    edge[b].append(a)
    
    
visited = [0] * (n+1)
parent = [0] * (n+1)
level = [0] * (n+1)

meas_level(1, 0)

m = int(input())
for i in range(m):
    a, b = map(int, input().split())
    
    print(lca(a, b))
```

```python
# 정점들의 거리

def meas_level(x, senior, dist):
    visited[x] = 1
    level[x] = level[senior] + 1
    
    if x != 1:
        distance[x] += distance[senior] + dist
    
    for i, j in edge[x]:
        if not visited[i]:
            parent[i][0] = x
            meas_level(i, x, j)
    
    
def lca(a, b):
    if level[b] < level[a]:
        a, b = b, a
        
    for i in range(17, -1, -1):
        if level[b] - level[a] >= 2 ** i:
            b = parent[b][i]
            
    if a == b:
        return a
    
    for i in range(17, -1, -1):
        if parent[a][i] != parent[b][i]:
            a = parent[a][i]
            b = parent[b][i]
        
        
    return parent[a][0]


n = int(input())

edge = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b, c = map(int, input().split())
    edge[a].append((b, c))
    edge[b].append((a, c))
    
    
visited = [0] * (n+1)
parent = [[0] * 18 for _ in range(n+1)] # log(2,100000)

distance = [0] * (n+1)
level = [0] * (n+1)
level[0] = -1

meas_level(1, 0, 0)

for i in range(1, 18): # 노드의 부모 관계 설정
    for j in range(1, n+1):
        parent[j][i] = parent[parent[j][i-1]][i-1]
        
        
m = int(input())
for i in range(m):
    a, b = map(int, input().split())
    # 최단거리 = distance[a] - distance[lca] + distance[b] - distance[lca]
    print(distance[a] + distance[b] - 2 * distance[lca(a, b)])
    
```

```python
# 도로 네트워크

from collections import deque
# dp [0] : parent, [1] : 최소, [2] : 최대

def meas_level():
    queue = deque([(1, 1)])
    visited = [0] * (n+1)
    visited[1] = 1
    
    while queue:
        x, depth = queue.popleft()
        
        for i, j in edge[x]:
            if not visited[i]:
                visited[i] = 1
                level[i] = depth + 1
                dp[i][0] = [x, j, j]
                queue.append((i, depth+1))
                
    
def lca(a, b):
    maxi = 0
    mini = int(1e20)
    
    if level[b] < level[a]:
        a, b = b, a
        
    for i in range(18, -1, -1):
        if level[b] - level[a] >= 2 ** i:
            b = dp[b][i][0]
            maxi = max(maxi, dp[b][i][2])
            mini = min(mini, dp[b][i][1])
            
            
    if a == b:
        return (mini, maxi)
    
    for i in range(17, -1, -1):
        if dp[a][i][0] != dp[b][i][0]:
            a = dp[a][i][0]
            b = dp[b][i][0]
            
            mini = min(mini, dp[a][i][1], dp[b][i][1])
            maxi = max(maxi, dp[a][i][2], dp[b][i][2])
            
        
    mini = min(mini, dp[a][0][1], dp[b][0][1])
    maxi = max(maxi, dp[a][0][2], dp[b][0][2])
    
    return (mini, maxi)


n = int(input())

edge = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b, c = map(int, input().split())
    edge[a].append((b, c))
    edge[b].append((a, c))
    
    
level = [0] * (n+1)

dp = [[[0,0,0] for _ in range(18)] for _ in range(n+1)]

meas_level()

for i in range(1, 18):
    for j in range(1, n+1):
        dp[j][i][0] = dp[dp[j][i-1][0]][i-1][0]
        dp[j][i][1] = min(dp[j][i-1][1], dp[dp[j][i-1][0]][i-1][1])
        dp[j][i][2] = min(dp[j][i-1][2], dp[dp[j][i-1][0]][i-1][2])
        

m = int(input())

for i in range(m):
    a, b = map(int, input().split())
    
    print(*lca(a, b))
    
```
