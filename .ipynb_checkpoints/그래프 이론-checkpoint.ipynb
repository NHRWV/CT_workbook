```python
# 백준 2252

from collections import deque

n, m = map(int, input().split())

order = [0] * (n+1)
compare = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    compare[a].append(b)
    order[b] += 1
    

def topology():
    result = []
    queue = deque()
    for i in range(1, n+1):
        if order[i] == 0:
            queue.append(i)
            
    while queue:
        x = queue.popleft()
        result.append(x)
        
        for i in compare[x]:
            order[i] -= 1
            if order[i] == 0:
                queue.append(i)
    
    for i in result:
        print(i, end = ' ')
        
topology()
```

```python
# 백준 1766

import heapq

n, m = map(int, input().split())

order = [0] * (n+1)
compare = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    compare[a].append(b)
    order[b] += 1

for i in compare:
    i.sort()


def topology():
    result = []
    queue = []
    for i in range(1, n+1):
        if order[i] == 0:
            queue.append(i)
     
    heapq.heapify(queue)
    while queue:
        x = heapq.heappop(queue)
        result.append(x)
        
        for i in compare[x]:
            order[i] -= 1
            if order[i] == 0:
                heapq.heappush(queue, i)
    
    for i in result:
        print(i, end = ' ')    
    

topology()
```

```python
# 백준 13905
# 이분탐색 & BFS로 풀었다.
# 최소스패닝트리로 푸는 방법은 이거 참고. https://www.acmicpc.net/source/30349848


from collections import deque        
        
        
n, m = map(int, input().split())
s, e = map(int, input().split())


edge = [[] for _ in range(n+1)]
for i in range(m):
    h1, h2, k = map(int, input().split())
    edge[h1].append((h2, k))
    edge[h2].append((h1, k))
    
def bfs(edge, mid):
    visited = [0] * (n+1)
    visited[s] = 1 
    queue = deque([s])
    
    while queue:
        x = queue.popleft()
        
        if x == e:
            return True
        
        for nxt, cost in edge[x]:
            if visited[nxt] == 0 and mid <= cost:
                visited[nxt] = 1
                queue.append(nxt)
                
    return False


start = 1
end = int(1e9)    
answer = 0
while start <= end:
    mid = (start + end) // 2
    
    if bfs(edge, mid):
        answer = mid
        start = mid + 1
        
    else:
        end = mid - 1
        
        
print(answer)
```

```python
# 백준 20010

from collections import deque

n, k = map(int, input().split())

edge = []
for i in range(k):
    a, b, c = map(int, input().split())
    edge.append((c, a, b))
    
edge.sort()

parent = [0] * n
for i in range(n):
    parent[i] = i
    

def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]

def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x < y:
        parent[y] = x
    else:
        parent[x] = y
      
    
    

lst = []
answer = 0
for cost, a, b in edge:
    if find_parent(a) != find_parent(b):
        union(a, b)
        answer += cost
        
        lst.append((cost, a, b))

course = [[] for _ in range(n)]
for i in lst:
    course[i[1]].append((i[2], i[0]))
    course[i[2]].append((i[1], i[0]))
    

node = [i for i in range(n)]
maxi = 0
for i in node:
    queue = deque([i])
    visited = [0] * n
    dist = [0] * n
    
    
    while queue:
        x = queue.popleft()
        visited[x] = 1
        
        for nxt, cost in course[x]:
            if visited[nxt] == 0:
                visited[nxt] = 1
                dist[nxt] = dist[x] + cost
                
                queue.append(nxt)
                
                
    maxi = max(maxi, max(dist))
            
        
print(answer)
print(maxi)
```

```python
# 백준 16562

n, m, k = map(int, input().split())
pay = list(map(int, input().split()))

parent = [0] * (n+1)
for i in range(1, n+1):
    parent[i] = i
    
def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]

def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x < y:
        parent[y] = x
    else:
        parent[x] = y
        
        

for i in range(m):
    v, w = map(int, input().split())
    union(v, w)
    

friend = set()
need = 0
for i in range(1, n+1):
    
    if i not in friend:
        cost = pay[i-1]
        our = [i]
        
        for j in range(1, n+1):
            if j != i:
                if find_parent(i) == find_parent(j):
                    our.append(j)
                    cost = min(cost, pay[j-1])
        
        need += cost
        friend.update(our)
        
    
    
if need <= k:
    print(need)
else:
    print("Oh no")
    

        
        
```

```python
# 백준 14595
# 오름차순으로 정렬시킨 다음에, 최근에 부쉈던 방 중에 가장 큰 방 번호를 기억해서 그것보다 작으면 그 다음 방부터 부수게 함.
# 이런 방법도 있더라....https://www.acmicpc.net/source/30601076

n = int(input())
m = int(input())

parent = [0] * (n+1)
for i in range(1, n+1):
    parent[i] = i
    
    
def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]

def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x < y:
        parent[x] = y
        
    else:
        parent[y] = x
    
    
    
cnt = 0
lst = []
for i in range(m):
    x, y = map(int, input().split())
    lst.append((x, y))
    
lst = sorted(lst, key = lambda x : (x[0], x[1]))

memo = 0
for s, e in lst:
    if s < memo:
        s = memo
        
    for j in range(s, e+1):
        if find_parent(j) != find_parent(e):
            union(j, e)
            cnt += 1
            
            memo = e
            
print(n - cnt)            
```

```python
# 백준 14595

n = int(input())
m = int(input())

cnt = 0
lst = []
for i in range(m):
    x, y = map(int, input().split())
    lst.append((x, y))
    
lst = sorted(lst, key = lambda x : (x[0], x[1]))

memo = 1
cnt = 1
for s, e in lst:
    if s <= memo:
        memo = max(memo, e)
        
    else:
        cnt += s - memo
        memo = e
        
cnt += n - memo
            
print(cnt)            
```

```python
# 백준 1516

from collections import deque


n = int(input())

time = [0] * (n+1)
order = [0] * (n+1)
edge = [[] for _ in range(n+1)]
for i in range(1, n+1):
    data = list(map(int, input().split()))[:-1]
    time[i] = data[0]
    for j in data[1:]:
        order[i] += 1
        edge[j].append(i)
    
    

def topology():
    result = time[:]
    queue = deque()
    for i in range(1, n+1):
        if order[i] == 0:
            queue.append(i)
            
    while queue:
        x = queue.popleft()
        
        for i in edge[x]:
            result[i] = max(result[i], result[x] + time[i])
            order[i] -= 1
            if order[i] == 0:
                queue.append(i)
    
    for i in range(1, n+1):
        print(result[i])
        
topology()
```

```python
# 백준 14567

from collections import deque

n, m = map(int, input().split())

time = [1] * (n+1)
order = [0] * (n+1)
edge = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    edge[a].append(b)
    order[b] += 1
    
def topology():
    result = time[:]
    queue = deque()
    for i in range(1, n+1):
        if order[i] == 0:
            queue.append(i)
    
    cnt = 1       
    while queue:
        x = queue.popleft()
        
        for i in edge[x]:
            order[i] -= 1
            if order[i] == 0:
                queue.append(i)
            
            if result[i] == 1:    
                result[i] += cnt 
                
        cnt += 1
                
    
    del result[0]
    print(*result)
    
    
topology()
```
