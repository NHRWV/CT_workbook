```python
# 트리


def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]


def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x < y:
        parent[y] = x
    else:
        parent[x] = y
        

n, q = map(int, input().split())

parent = [i for i in range(n+1)]

edge = [1] * (n+1)
for i in range(2, n+1):
    a = int(input())
    edge[i] = a
    

q += n-1
query = [[0, 0] for _ in range(q)]

for i in range(q):
    oper = list(map(int, input().split()))
    
    if not oper[0]:
        query[i][0] = oper[1]
    else:
        query[i][0] = oper[1]
        query[i][1] = oper[2]
        
        
answer = []
for i in range(q-1, -1, -1):
    if not query[i][1]:
        union(query[i][0], edge[query[i][0]])
        
    else:
        if find_parent(query[i][0]) == find_parent(query[i][1]):
            answer.append(1)
        else:
            answer.append(0)
            
            
while answer:
    if answer.pop():
        print('YES')
    else:
        print('NO')
            
```

```python
# 수열과 쿼리 22

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    
    
    
n = int(input())
seq = list(map(int, input().split()))
m = int(input())

query1 = []
query2 = []
for i in range(m):
    oper = list(map(int, input().split()))
    
    if oper[0] == 1:
        query1.append(oper[1:])
    else:
        temp = oper[1:] + [i]
        query2.append(temp)
        
        
query2.sort()

func = lambda a, b: a + b
seg = SegmentTree(seq, 0, func = func)

tag = 0
answer = []
for k, i, j, idx in query2:
    if tag == k:
        answer.append((idx, seg.query(i-1, j-1)))
        
    else:
        for p in range(tag+1, k+1):
            a, b = query1[p-1]
            seg[a-1] = b
            
        tag = k    
        answer.append((idx, seg.query(i-1, j-1)))
        
        
answer.sort()
for i, j in answer:
    print(j)
        
```

```python
# MooTube (Gold)


import sys

def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]


def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x < y:
        parent[y] = x
    else:
        parent[x] = y


n, q = map(int, input().split())

edge = []
for i in range(n-1):
    a, b, r = map(int, input().split())
    edge.append((r, a, b))

edge.sort(reverse = True)

parent = [i for i in range(n+1)]
parent_num = [1] * (n+1)


query = []
for i in range(q):
    k, v = map(int, input().split())
    query.append((k, v, i))
    
query.sort(reverse = True)
    

answer = [] 
i = 0

for k, v, idx in query:
    while i <= n - 2 and edge[i][0] >= k:
        x = edge[i][1]
        y = edge[i][2]
        
        hab = parent_num[find_parent(x)] + parent_num[find_parent(y)]
        union(x, y)
        parent_num[find_parent(x)] = hab
        i += 1
        
    answer.append((idx, parent_num[find_parent(v)] - 1))
#    print(v)
#    print(find_parent(v))
#    print(parent_num[find_parent(v)] - 1)
#    print()
    
#print()
answer.sort()
for i, j in answer:
    print(j)
    
```

https://www.acmicpc.net/problem/12012

```python
# Closing the Farm (Gold)

def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]


def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x == y:
        return False
    
    if x < y:
        parent[y] = x
    else:
        parent[x] = y
    
    return True
        

n, m = map(int, input().split())

parent = [i for i in range(n+1)]

edge = [[] for _ in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    edge[a].append(b)
    edge[b].append(a)
    

query = []
for i in range(n):
    q = int(input())
    
    query.append(q)
       

visited = [0] * (n+1)
        
answer = []
check = 0
for i in range(n-1, -1, -1):
    check += 1
    cur = query[i]
    for j in edge[cur]:
        if visited[j] and union(cur, j):
            check -= 1
            
    visited[cur] = 1
    answer.append(check)
    
            
#print(answer)            
while answer:
    if answer.pop() == 1:
        print('YES')
    else:
        print('NO')
            
```
