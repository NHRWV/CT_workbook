https://www.acmicpc.net/problem/18809

```python
# Gaaaaaaaaaarden

from collections import deque
from itertools import combinations


n, m, g, r = map(int, input().split())

garden = []
psb = []

for i in range(n):
    a = list(map(int, input().split()))
    garden.append(a)
    for j in range(m):
        if a[j] == 2:
            psb.append((i, j))
            
maxi = 0

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]


def sol(red, green):
    rqueue = deque()
    gqueue = deque()
    
    count = 1
    flower = 0
    
    for i, j in red:
        visited[i][j] = -1
        rqueue.append((i, j))
    for i, j in green:
        visited[i][j] = -1
        gqueue.append((i, j))
        
    while rqueue and gqueue:
        for i in range(len(rqueue)):
            x, y = rqueue.popleft()
            
            if visited[x][y] == -2:
                continue
                
            for j in range(4):
                nx = x + dx[j]
                ny = y + dy[j]
                
                if nx < 0 or nx >= n or ny < 0 or ny >= m:
                    continue
                    
                if garden[nx][ny] and not visited[nx][ny]:
                    visited[nx][ny] = count
                    rqueue.append((nx, ny))
                    
        for i in range(len(gqueue)):
            x, y = gqueue.popleft()
            
            if visited[x][y] == -2:
                continue
                
            for j in range(4):
                nx = x + dx[j]
                ny = y + dy[j]
                
                if nx < 0 or nx >= n or ny < 0 or ny >= m:
                    continue
                    
                if garden[nx][ny]:
                    if not visited[nx][ny]:
                        visited[nx][ny] = -1
                        gqueue.append((nx, ny))
                        
                    elif visited[nx][ny] == count:
                        flower += 1
                        visited[nx][ny] = -2
                        
        count += 1
        
    return flower
            
        

for combo in combinations(psb, g+r):
    for green in combinations(combo, g):
        red = [x for x in combo if x not in green]
        visited = [[0] * m for _ in range(n)]
        
        result = sol(red, green)
        maxi = max(maxi, result)
        
print(maxi)
```

```python
# 소년 점프

from collections import deque

r, c = map(int, input().split())

maze = []
for i in range(r):
    maze.append(list(map(int, input().rstrip())))
    
visited = [[[-1] * 3 for _ in range(c)] for _ in range(r)]
queue = deque()

for i in range(3):
    x, y = map(int, input().split())
    
    visited[x-1][y-1][i] = 0
    queue.append((x-1, y-1, i))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]


while queue:
    x, y, idx = queue.popleft()
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if nx < 0 or nx >= r or ny < 0 or ny >= c:
            continue
            
        if not maze[nx][ny] and visited[nx][ny][idx] == -1:
            queue.append((nx, ny, idx))
            visited[nx][ny][idx] = visited[x][y][idx] + 1
            
            
count = 0            
answer = int(1e9)

for i in range(r):
    for j in range(c):
        if min(visited[i][j]) != -1:
            temp = max(visited[i][j])
            
            if answer == temp:
                count +=1
            
            elif answer > temp:
                answer = temp
                count = 1
            

if count:
    print(answer)
    print(count)
else:
    print(-1)
```

https://www.acmicpc.net/problem/1926

```python
# 그림

from collections import deque

n, m = map(int, input().split())

info = []
for i in range(n):
    info.append(list(map(int, input().split())))
    
count = 0

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def sol(x, y):
    if not info[x][y]:
        return False
    
    info[x][y] = 0
    queue = deque([(x, y)])
    
    global cnt
    cnt = 1
    
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
                
            if info[nx][ny]:
                info[nx][ny] = 0
                queue.append((nx, ny))
                cnt += 1
                
    return True


maxi_cnt = 0
for i in range(n):
    for j in range(m):
        if sol(i, j):
            count += 1
            maxi_cnt = max(maxi_cnt, cnt)
            
            
print(count)
print(maxi_cnt)
```

https://www.acmicpc.net/problem/1743

```python
# 음식물 피하기

from collections import deque

n, m, k = map(int, input().split())

space = [[0] * m for _ in range(n)]
for i in range(k):
    r, c = map(int, input().split())
    space[r-1][c-1] = 1
    

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def sol(x, y):
    if not space[x][y]:
        return False
    
    space[x][y] = 0
    queue = deque([(x, y)])
    
    global size
    size = 1
    
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
                
            if space[nx][ny]:
                space[nx][ny] = 0
                size += 1
                queue.append((nx, ny))
                
    return True


answer = 0
for i in range(n):
    for j in range(m):
        if sol(i, j):
            answer = max(answer, size)
            
            
print(answer)
                
```

```python
# 숨바꼭질 4

from collections import deque

n, k = map(int, input().split())
lst = [-1] * 100001

path = []
queue = deque([(n, 0)])
lst[n] = 0



while queue:
    x, cnt = queue.popleft()
    
    if x == k:
        while x != n:
            path.append(x)
            x = lst[x]
        
        path.append(x)
        print(cnt)
        print(*path[::-1])
        break
        
    dx = [x - 1, x + 1, 2 * x]
    
    for i in range(3):
        nx = dx[i]
        
        if nx < 0 or nx >= 100001:
            continue
            
        if lst[nx] == -1:
            lst[nx] = x
            queue.append((nx, cnt+1))
    
```

https://www.acmicpc.net/problem/16953

```python
# A → B

from collections import deque
import sys

a, b = map(int, input().split())


queue = deque([(a, 0)])


while queue:
    x, cnt = queue.popleft()
    
    if x == b:
        print(cnt + 1)
        sys.exit()
        
    dx = [2 * x, int(str(x) + '1')]
    
    for i in range(2):
        nx = dx[i]
        
        if nx > b:
            continue
            
        
        
        queue.append((nx, cnt+1))
            
            
print(-1)
```

https://www.acmicpc.net/problem/1937

```python
# 욕심쟁이 판다


n = int(input())

forest = []
for i in range(n):
    forest.append(list(map(int, input().split())))
    

dp = [[0] * n for _ in range(n)]    
    
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def sol(x, y):
    if not dp[x][y]:
        dp[x][y] = 1
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= n:
                continue
                
            if forest[nx][ny] > forest[x][y]:
                dp[x][y] = max(dp[x][y], sol(nx, ny) + 1)
                
    return dp[x][y]


answer = 0

for i in range(n):
    for j in range(n):
        answer = max(answer, sol(i, j))
        
print(answer)

```
